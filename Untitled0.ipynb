```python
%%writefile app.py

import streamlit as st
from graphviz import Digraph

# ---------- App Title ----------
st.title("ðŸ¤– AI-Powered Requirements-to-Code Generator")
st.write("Enter software requirements and generate UML, ERD, and Python code skeleton.")

# ---------- User Input ----------
requirements = st.text_area(
    "ðŸ“Œ Enter your software requirements:",
    placeholder="Example: I want a student management system with login, database, and course registration"
)

# ---------- AI Logic (Rule-Based NLP) ----------
def generate_output(req):
    req = req.lower()

    # UML Use Case
    use_cases = []
    if "login" in req:
        use_cases.append("User Login")
    if "register" in req:
        use_cases.append("User Registration")
    if "student" in req:
        use_cases.append("Manage Student")
    if "course" in req:
        use_cases.append("Manage Course")
    if "database" in req:
        use_cases.append("Store Data")

    # ERD Entities
    entities = []
    if "student" in req:
        entities.append("Student (id, name, email)")
    if "course" in req:
        entities.append("Course (course_id, title)")
    if "user" in req or "login" in req:
        entities.append("User (user_id, username, password)")

    # Python Code Skeleton
    code = """# Auto-generated Python Code Skeleton

class User:
    def login(self):
        pass

class Student:
    def add_student(self):
        pass

class Course:
    def add_course(self):
        pass
"""

    return use_cases, entities, code

# ---------- Diagram Generators ----------
def draw_use_case_diagram(use_cases):
    dot = Digraph(comment='Use Case Diagram', format='png')
    dot.attr('node', shape='ellipse', style='filled', color='lightblue')

    dot.node('Actor', 'User')

    for uc in use_cases:
        dot.node(uc)
        dot.edge('Actor', uc)

    return dot

def draw_erd(entities):
    dot = Digraph(comment='ER Diagram', format='png')
    dot.attr('node', shape='record', style='filled', color='lightyellow')

    for e in entities:
        # Split entity name and attributes
        name_attr = e.split('(')
        entity_name = name_attr[0].strip()
        attrs = name_attr[1].replace(')', '').strip() if len(name_attr) > 1 else ''
        dot.node(entity_name, f"{{{entity_name}|{attrs}}}")

    return dot

# ---------- Button ----------
if st.button("ðŸš€ Generate Output"):
    if requirements.strip() == "":
        st.warning("Please enter some requirements first.")
    else:
        use_cases, entities, code = generate_output(requirements)

        # Graphical Use Case
        st.subheader("ðŸ“Š UML Use Case Diagram (Graphical)")
        uc_dot = draw_use_case_diagram(use_cases)
        st.graphviz_chart(uc_dot)

        # Graphical ERD
        st.subheader("ðŸ—„ ER Diagram (Graphical)")
        er_dot = draw_erd(entities)
        st.graphviz_chart(er_dot)

        # Python Code Skeleton
        st.subheader("ðŸ’» Python Code Skeleton")
        st.code(code, language="python")
```
